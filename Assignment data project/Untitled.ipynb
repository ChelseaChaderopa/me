```python
cvc = Fastfood_data["city"].value_counts()
cvc[cvc > 40].plot(kind="bar")
plt.title("Restaurant locations")
plt.ylabel("number of restaurants")
plt.xlabel("city")
```

```python
restaurants_counts = Fastfood_data.name.value_counts()
restaurants_counts = restaurants_counts[restaurants_counts>150]
restaurants_list = list(restaurants_counts.index)
restaurants_list
```

```python
data_top = Fastfood_data[Fastfood_data.name.isin(restaurants_list)]
data_top = data_top[["country", "province", "city", "name"]]
```

```python
data_top_province = data_top.province.value_counts()
```

```python
data_top_province_list = list(data_top_province[data_top_province>250].index)
```

```python
data_top_province_top = data_top[data_top.province.isin(data_top_province_list)]
```

```python
data_top_province_top_group = data_top_province_top.groupby(["province","name"])
data_top_province_top_group.size().unstack().plot(kind ="bar", figsize =(16,7))
```
